```python
import pandas as pd, json
```

```python
samples_frame = pd.read_csv('Samples-Of-Interest.csv')
```

```python
samples_json = json.loads(samples_frame['Sample ID'].to_json())
```

```python
list_of_samples_filenames = [('sample-' + str(int(v[7:])) + '.csv', v) for k,v in samples_json.items()]
print(list_of_samples_filenames[0])
```

```python
single_test_frame = pd.read_csv('blood-samples/sample-1001.csv')
```

```python
cols1 = single_test_frame.columns.tolist()
cols2 = samples_frame.columns.tolist()
cols1.extend(cols2)

newdf = pd.DataFrame(columns=cols1)
print(newdf.columns)
```

```python
for filename, sample_name in list_of_samples_filenames:
    print(filename, sample_name)
    sample_df = pd.read_csv('blood-samples/' + filename)
    this_sample_metadata = samples_frame[samples_frame['Sample ID'] == sample_name]
    display(this_sample_metadata)
    
    sample_df['Sample ID'] = this_sample_metadata.iloc[0]['Sample ID']
    sample_df['Biosample type'] = this_sample_metadata.iloc[0]['Biosample type']
    sample_df['Biosample name'] = this_sample_metadata.iloc[0]['Biosample name']
    sample_df['Tissue type'] = this_sample_metadata.iloc[0]['Tissue type']
    sample_df['Cancer type'] = this_sample_metadata.iloc[0]['Cancer type']
    sample_df['Region number'] = this_sample_metadata.iloc[0]['Region number']
    newdf = pd.concat([newdf, sample_df])
```

```python
display(newdf)
```

```python
newdf_clean = newdf.drop('Unnamed: 0', axis=1)
```

```python
display(newdf_clean)
```
